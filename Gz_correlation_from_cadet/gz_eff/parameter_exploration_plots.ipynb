### t_breakthrough
When c_out/c_feed = 0.01

```python
results = []

for file in sims:
    try:
        c_load, this_dp, ka, q_max, kd, keq, t, c_out, fVolum = explor.get_sim_results(results_folder, file, True)
        if dp == this_dp and fVolum == 1.633e-8/4:
            norm_c = c_out/c_load
            idx = np.searchsorted(norm_c, 0.01, side="left")
            results.append([c_load, keq, t[idx], q_max, dp])
    except:
        pass
```

```python
df = pd.DataFrame(results, columns=['c', 'keq', 't_break', 'q_max', 'dp'])

c_set = list(set(df['c']))
c_set.sort()
c_set.reverse()

q_max = df.at[0, 'q_max']
c_set
```

```python
params = {'font.weight':'normal', 'font.size':20, 'figure.autolayout':True}
plt.rcParams.update(params)
fig, ax = plt.subplots()
ax.set_ylabel(r't$_{breakthrough}$ [s]')
ax.set_xlabel(r'K$_{eq}$ [-]')
ax.set_xscale('log')
ax.set_yscale('log')
fig.patch.set_facecolor('white')
fig.set_size_inches(8, 6, forward=True)

keq_exreme = [min(df['keq']), max(df['keq'])]
keq_vals = np.linspace(keq_exreme[0], keq_exreme[1], 5000)
tr_vals  = [explor.get_t_r(keq) for keq in keq_vals]
t0 = explor.get_t0()
t0_min = explor.get_min_t0()

ax.plot(keq_vals, tr_vals, color='black')
ax.plot(keq_exreme, [t0, t0], color='black')
ax.plot(keq_exreme, [t0_min, t0_min], color='black')

for i, c in enumerate(c_set):
    df_temp = df[df['c'] == c].copy()
    df_temp.sort_values(by=['keq'], inplace=True)
    df_temp.reset_index(inplace=True, drop=True)

    ax.scatter(df_temp['keq'], df_temp['t_break'], label=f'c = {c*mass:.0e} mg/ml')
    
    keq_lin = explor.get_keq_lin(q_max, c, 0.99)
    if keq_lin < keq_exreme[1]:
        idx = df_temp['keq'].searchsorted(keq_lin, side='left')
        try:
            ax.scatter(df_temp.at[idx-1, 'keq'], df_temp.at[idx-1, 't_break'], 
                      marker='|', s=700, color=colors[i])
        except ValueError:
            pass
        except KeyError:
            pass
    
    keq_cross = explor.get_keq_tr_tsat(q_max, c)
    if keq_cross < max(df['keq']):
        t_sat = explor.get_t_saturation(c, q_max)
        ax.plot([keq_cross, max(df['keq'])], [t_sat, t_sat]) # color=plt.gca().lines[-1].get_color()

ax.legend(loc='upper left', frameon=False, handlelength=1.0, fontsize=15)
# fig.savefig(os.path.join(image_folder, f't_break_vs_Keq_0.01_log_dp_{dp:.2e}.png'), dpi=300)
```

## Dimensionless (i.e. better) mapping slope vs Dp and flow rate
Note I'm changing the folders below

```python
def get_linear_v(Q, D=5.0e-3):
    return Q/(np.pi/4.0*D**2.0)

def get_t_star(t, v_lin, l_col):
    return t*v_lin/l_col

def get_peclet(d_part, v_lin, Dp, eps_c):
    return d_part*v_lin/Dp/eps_c
```

```python
d_part = 5.0e-5
l_col  = 0.042

eps_t = 0.694
up_bnd = (1.0 - eps_t)
up_bnd

eps_c = 0.49

V_col = np.pi/4 * (0.5**2) * 1e-4 * 0.042

```

```python
slope_res = []

for i, cont in df_lo.iterrows():
    dp = cont['dp']
    fVolum = cont['fVolum']
    v_lin = get_linear_v(fVolum, D=5.0e-3)
    t0 = V_col*eps_c/fVolum
    t0_star = get_t_star(t0, v_lin, l_col)
    
    df_temp = df_hi[(df_hi['dp'] == dp) & (df_hi['fVolum'] == fVolum)].copy()
    df_temp.reset_index(drop=True, inplace=True)
    
    t_lo = cont['t_break_01']
    t_hi = df_temp.at[0, 't_break_01']
    t_star_lo = get_t_star(t_lo, v_lin, l_col)
    t_star_hi = get_t_star(t_hi, v_lin, l_col)
#     slope_1 = (t_star_hi - t_star_lo)/(1000.0 - 1.0) # Change denominator if using different Keq
    slope_1 = (t_star_hi - t0_star)/(1000.0)
    
    t_lo = cont['t_break_50']
    t_hi = df_temp.at[0, 't_break_50']
    t_star_lo = get_t_star(t_lo, v_lin, l_col)
    t_star_hi = get_t_star(t_hi, v_lin, l_col)
#     slope_2 = (t_star_hi - t_star_lo)/(1000.0 - 1.0) # Change denominator if using different Keq
    slope_2 = (t_star_hi - t0_star)/(1000.0)
    
    pe = get_peclet(d_part, v_lin, dp, eps_c)
    slope_res.append([pe, slope_1, slope_2])

df_slope = pd.DataFrame(slope_res, columns=['pe', 'slope_01', 'slope_50'])
```

```python
def plot_slope_df(df, frac_bnd=(0, 1)):   
    x = np.array(df.pop('pe'))

    for i in df.columns:
        if i < frac_bnd[0] or i > frac_bnd[1]:
            df.drop(labels=i, axis=1, inplace=True)
    
    y = []
    for col in df.columns:
        y.append(float(col))

    X, Y = np.meshgrid(x, y)
    X = X.transpose()
    Y = Y.transpose()

    Z = np.zeros((len(x), len(y)))
    for row in range(len(x)):
        for col in range(len(y)):
            Z[row, col] = df.values[row, col]/(1.0 - eps_t)

    fig = plt.figure()
    fig.set_size_inches(15, 12, forward=True)
    ax = fig.add_subplot(111, projection='3d')
    params = {'font.weight':'normal', 'font.size':20, 'figure.autolayout':True}
    plt.rcParams.update(params)
    fig.patch.set_facecolor('white')
    surf = ax.plot_surface(X, Y, Z, cmap=cm.plasma)
#     fig.colorbar(surf, shrink=0.5) # shrink=0.5, aspect=5

    pad = 15
    ax.set_xlabel(r'$P\'{e}$ [-]', labelpad=pad)
    ax.set_ylabel(r'c$_{out}$/c$_{load}$ [-]', labelpad=pad)
    ax.set_zlabel(r'$\frac{\Delta \; CV}{\Delta \; K_{eq} \; (1 - \epsilon_t)}$ [-]', labelpad=pad)

    tick_size = 14
    ax.tick_params(labelsize=tick_size)
    ax.view_init(10, 290)
    ax.dist = 10.5
    
    return fig, ax
```

```python
res = []

for i, cont in df_hi.iterrows():
    dp = cont['dp']
    fVolum = cont['fVolum']
    v_lin = get_linear_v(fVolum, D=5.0e-3)
    pe = get_peclet(d_part, v_lin, dp, eps_c)
    
    t0 = V_col*eps_c/fVolum
    t0_star = get_t_star(t0, v_lin, l_col)
    
    these_res = [pe]
    for frac in recovered_fracs:
        t_hi = cont[frac]
        t_star_hi = get_t_star(t_hi, v_lin, l_col)
        slope = (t_star_hi - t0_star)/cont['keq']
        these_res.append(slope)
    
    res.append(these_res)

df_slope = pd.DataFrame(res, columns=['pe'] + [round(i, 2) for i in fracs])
```

```python
profiles_1000 = {} # Keyed by Pe
profiles_1 = {} # Keyed by Pe

for file in sims:
    c_load, this_dp, ka, q_max, kd, keq, t, c_out, fVolum =\
    explor.get_sim_results(results_folder, file, True)
    
    df_results = pd.DataFrame(np.array([t, c_out/c_load]).transpose(), columns=['time_s', 'c_normal'])
    df_results['CV'] = fVolum*df_results['time_s']/V_col
    v_lin = explor.get_linear_v(fVolum, D=5.0e-3)
    pe = get_peclet(d_part, v_lin, this_dp, eps_c)
    
    if round(keq) == 1000:
        profiles_1000[pe] = df_results
    elif round(keq) == 1:
        profiles_1[pe] = df_results
```
