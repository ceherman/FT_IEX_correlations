## Lyszome dilute breakthrough, 20% B, rep 2

```python
directory = './data/randomized_data_20_B_rep_2/'
files = os.listdir(directory)
files
```

```python
my_data = {}
for f in files:
    my_data[f] = my_plot.data(directory, f)
```

```python
velocities = []

for i, f in enumerate(files):
    if i == 0 or i == 2:
        ind = 6000
    else:
        ind = 3000
        
    my_data[f].flow = my_data[f].df.at[ind, 'P960_Flow_ml/min']

    my_data[f].v_col = 2.0
    my_data[f].velocity = my_data[f].flow/(np.pi/4.0*0.5**2)*60.0

    velocities.append(my_data[f].velocity)

print(velocities)
velocities.sort()
print(velocities)
```

```python
fig, ax = plt.subplots()
font    = {'weight' : 'normal', 'size'   : 20}
plt.rc('font', **font)
fig.set_size_inches(8, 6, forward=True)
ax.set_xlabel('Column volumes [-]')
ax.set_ylabel('A215 [mAU]')
ax2 = ax.twinx()
ax2.set_ylabel('Flow [ml/min]')

for v in velocities:
    for f in files:
        if my_data[f].velocity == v:
            df = my_data[f].df
            inject_cv = [x/my_data[f].v_col for x in df['Inject_ml'] if str(x) != 'nan']
            ax.plot(df['UV1_215nm_ml']/my_data[f].v_col - inject_cv[-1], df['UV1_215nm_mAU'], label=f'{int(my_data[f].velocity)} cm/h')
            ax2.plot(df['P960_Flow_ml']/my_data[f].v_col - inject_cv[-1], df['P960_Flow_ml/min'], '--')

ax.legend(loc='lower right', frameon=False, handlelength=1)
ax.set_xlim(0, 20)
ax.set_ylim(-10, 300)
ax2.set_ylim(0, 1.2)

my_plot.align_yaxis_multiple([ax, ax2])

fig.patch.set_alpha(1.0)
plt.tight_layout()
# fig.savefig('./images/20_B_flow_rate_comparison_zoomed.png', dpi=300)




```

```python
fig, ax = plt.subplots()
font    = {'weight' : 'normal', 'size'   : 20}
plt.rc('font', **font)
fig.set_size_inches(8, 6, forward=True)
ax.set_xlabel('Column volumes [-]')
ax.set_ylabel('A215 [mAU]')

for v in velocities:
    for f in files:
        if my_data[f].velocity == v:
            df = my_data[f].df
            inject_cv = [x/my_data[f].v_col for x in df['Inject_ml'] if str(x) != 'nan']
            ax.plot(df['UV1_215nm_ml']/my_data[f].v_col - inject_cv[-1], df['UV1_215nm_mAU'], label=f'{int(my_data[f].velocity)} cm/h')

ax.legend(loc='lower right', frameon=False, handlelength=1)
ax.set_xlim(0, 20)
ax.set_ylim(-10, 250)

fig.patch.set_alpha(1.0)
plt.tight_layout()
# fig.savefig('./images/20_B_randomized_study_rep_2.png', dpi=300)
```

# Blue dextran pulses

```python
directory = './data_blue_dextran_on_column/'
files = os.listdir(directory)

image_directory = './images/'
try:
    os.mkdir(image_directory)
except:
    pass

files
```

```python
data_break = {}

for f in files:
    data_break[f] = data(directory, f)
```

```python
for f in files:
    if 'P960_Flow_ml/min' in data_break[f].df.columns or 'Flow_ml/min' in data_break[f].df.columns:
        if 'P960_Flow_ml/min' in data_break[f].df.columns and data_break[f].df.at[100, 'P960_Flow_ml/min'] > 0:
            data_break[f].flow = data_break[f].df.at[500, 'P960_Flow_ml/min']
        elif 'Flow_ml/min' in data_break[f].df.columns and data_break[f].df.at[100, 'Flow_ml/min'] > 0:
            data_break[f].flow = data_break[f].df.at[500, 'Flow_ml/min']

        data_break[f].v_col = 2.0
        data_break[f].velocity = data_break[f].flow/(np.pi/4.0*0.5**2)*60.0

        print(f, data_break[f].velocity)
        
        for col_name in data_break[f].df.columns:
            if 'min' in col_name:
                data_break[f].df[col_name[:-4] + '_cv'] = data_break[f].df[col_name]*data_break[f].flow/data_break[f].v_col
            if 'ml' in col_name:
                data_break[f].df[col_name[:-3] + '_cv'] = data_break[f].df[col_name]/data_break[f].v_col
    else:
        print(f)
```

```python
fig, ax = plt.subplots()
font    = {'weight' : 'normal', 'size'   : 20}
plt.rc('font', **font)
fig.set_size_inches(8, 6, forward=True)
ax.set_xlabel('Column volumes [-]')
ax.set_ylabel('A280 [mAU]')

for i, f in enumerate(files):
    df = data_break[f].df
    ax.plot(df.UV2_280nm_cv, df.UV2_280nm_mAU, label=f'Rep {i+1}')

# ax.set_xlim(0, 3)
ax.legend(loc='best', frameon=False, handlelength=1)
fig.patch.set_alpha(1.0)
plt.tight_layout()
# fig.savefig('./images/blue_dextran.png', dpi=300)
```

```python
for f in files:
    if 'P960_Flow_ml/min' in data_break[f].df.columns or 'Flow_ml/min' in data_break[f].df.columns:
        if 'P960_Flow_ml/min' in data_break[f].df.columns and data_break[f].df.at[100, 'P960_Flow_ml/min'] > 0:
            data_break[f].flow = data_break[f].df.at[500, 'P960_Flow_ml/min']
        elif 'Flow_ml/min' in data_break[f].df.columns and data_break[f].df.at[100, 'Flow_ml/min'] > 0:
            data_break[f].flow = data_break[f].df.at[500, 'Flow_ml/min']

        data_break[f].v_col = 2.0
        data_break[f].velocity = data_break[f].flow/(np.pi/4.0*0.5**2)*60.0

        print(f, data_break[f].velocity)
        
        for col_name in data_break[f].df.columns:
            if 'min' in col_name:
                data_break[f].df[col_name[:-4] + '_cv'] = data_break[f].df[col_name]*data_break[f].flow/data_break[f].v_col
            if 'ml' in col_name:
                data_break[f].df[col_name[:-3] + '_cv'] = data_break[f].df[col_name]/data_break[f].v_col
    else:
        print(f)
```

```python
rep_dict = {"a":1, "b":2, "c":3}

for i, f in enumerate(files):    
    letter = f[f.find('.')-1]
    print(letter, rep_dict[letter])
```

```python
fig, ax = plt.subplots()
font    = {'weight' : 'normal', 'size'   : 20}
plt.rc('font', **font)
fig.set_size_inches(8, 6, forward=True)
ax.set_xlabel('Column volumes [-]')
ax.set_ylabel('A280 [mAU]')

for i, f in enumerate(files):    
    df = data_break[f].df
    letter = f[f.find('.')-1]
    rep_num = rep_dict[letter]
    label = f"{int(data_break[f].velocity)} cm/h, rep {rep_num}"
    
    ax.plot(df.UV2_280nm_cv, df.UV2_280nm_mAU, label=label)

ax.set_xlim(0, 3)
ax.legend(loc='best', frameon=False, handlelength=1)
fig.patch.set_alpha(1.0)
plt.tight_layout()
# fig.savefig('./images/lysozyme_on_column.png', dpi=300)
```
